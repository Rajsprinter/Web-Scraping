```python
# Import necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# Q1: Scrape data from Naukri.com for "Data Scientist" jobs in "Delhi/NCR" with a salary filter of "3-6 Lakhs"

# Initialize the Chrome WebDriver for Naukri.com
driver = webdriver.Chrome()

# Open the Naukri website
driver.get("https://www.naukri.com/")

# Wait for the page to load
time.sleep(3)

# Locate the search field for "Skills, Designations, Companies" and enter "Data Scientist"
search_field = driver.find_element(By.CSS_SELECTOR, "input.suggestor-input")
search_field.send_keys("Data Scientist")

# Press ENTER to search
search_field.send_keys(Keys.RETURN)

# Wait for the search results to load
time.sleep(3)

# Apply the location filter for "Delhi/NCR"
location_filter = driver.find_element(By.XPATH, "//span[text()='Delhi / NCR']")
location_filter.click()

# Wait for the filter to be applied
time.sleep(2)

# Apply the salary filter for "3-6 Lakhs"
salary_filter = driver.find_element(By.XPATH, "//span[text()='3-6 Lakhs']")
salary_filter.click()

# Wait for the filter to be applied
time.sleep(3)

# Scrape the first 10 job results - title, location, company name, experience required
job_titles = []
locations = []
companies = []
experiences = []

# Locate the job listing container and extract the first 10 jobs
job_cards = driver.find_elements(By.CSS_SELECTOR, ".jobTuple")[:10]

for job in job_cards:
    # Extract job title
    job_title = job.find_element(By.CSS_SELECTOR, "a.title").text
    job_titles.append(job_title)
    
    # Extract job location
    location = job.find_element(By.CSS_SELECTOR, ".location").text
    locations.append(location)
    
    # Extract company name
    company = job.find_element(By.CSS_SELECTOR, ".companyInfo .subTitle").text
    companies.append(company)
    
    # Extract experience required
    experience = job.find_element(By.CSS_SELECTOR, ".experience").text
    experiences.append(experience)

# Creating a DataFrame for the scraped data
df_naukri = pd.DataFrame({
    "Job Title": job_titles,
    "Location": locations,
    "Company": companies,
    "Experience Required": experiences
})

# Close the driver for Naukri
driver.quit()

# Display the DataFrame
print("Naukri Data (Delhi/NCR, 3-6 Lakhs):")
df_naukri


# Q2: Scrape data from Shine.com for "Data Analyst" jobs in "Bangalore"

# Initialize the Chrome WebDriver for Shine.com
driver = webdriver.Chrome()

# Open the Shine website
driver.get("https://www.shine.com/")

# Wait for the page to load
time.sleep(3)

# Locate the search field for "Job title, Skills" and enter "Data Scientist"
search_field_title = driver.find_element(By.CSS_SELECTOR, "input[placeholder='Job title, Skills']")
search_field_title.send_keys("Data Scientist")

# Locate the search field for "enter the location" and enter "Bangalore"
search_field_location = driver.find_element(By.CSS_SELECTOR, "input[placeholder='Location']")
search_field_location.send_keys("Bangalore")

# Click the search button
search_button = driver.find_element(By.CSS_SELECTOR, "button.search-btn")
search_button.click()

# Wait for the search results to load
time.sleep(3)

# Scrape the first 10 job results - title, location, company name, experience required
job_titles_shine = []
locations_shine = []
companies_shine = []
experiences_shine = []

# Locate the job listing container and extract the first 10 jobs
job_cards_shine = driver.find_elements(By.CSS_SELECTOR, ".w-90")[:10]

for job in job_cards_shine:
    # Extract job title
    job_title = job.find_element(By.CSS_SELECTOR, "span.d-block").text
    job_titles_shine.append(job_title)
    
    # Extract job location
    location = job.find_element(By.CSS_SELECTOR, ".loc").text
    locations_shine.append(location)
    
    # Extract company name
    company = job.find_element(By.CSS_SELECTOR, ".d-flex.flex-column .mt-2").text
    companies_shine.append(company)
    
    # Extract experience required (optional as some jobs may not show this clearly)
    try:
        experience = job.find_element(By.CSS_SELECTOR, ".w-80 .exp").text
    except:
        experience = "Not specified"
    experiences_shine.append(experience)

# Creating a DataFrame for the scraped data from Shine
df_shine = pd.DataFrame({
    "Job Title": job_titles_shine,
    "Location": locations_shine,
    "Company": companies_shine,
    "Experience Required": experiences_shine
})

# Close the driver for Shine
driver.quit()

# Display the DataFrame
print("Shine Data (Bangalore):")
df_shine
```
